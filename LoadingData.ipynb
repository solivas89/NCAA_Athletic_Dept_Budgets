```python
import pandas as pd
import numpy as np
import pymongo
```

```python
excel_file1 = "Resources/data/NBA from Wikipedia.xlsx"
excel_file2 = "Resources/data/Project 2 - Mapping of NBA Roster to College File.xlsx"
excel_file3 = "Resources/data/Schools - Budget Data.xlsx"
excel_file4 = "Resources/data/Tax Data - Re-Formatted.xlsx"
df1 = pd.read_excel(excel_file1)
df2 = pd.read_excel(excel_file2)
df3 = pd.read_excel(excel_file3)
df4 = pd.read_excel(excel_file4)
```

```python
# Initialize PyMongo to work with MongoDB
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define database
db = client.NCAA_Athletics
```

```python
# Define NBA Wikipedia Collection
collection = db['NBA_Wikipedia']

# Convert it to a dictionary
NBAWiki_dict = df1.to_dict('records')
# print(NBAWiki_dict)
# Inserting the dictionary into collection
collection.insert_many(NBAWiki_dict)
```

```python
# Define NBA From College Collection
collection = db['NBA_From_College']

# Convert it to a dictionary
NBA_College = df2.to_dict('records')
# print(NBA_College)
# Inserting the dictionary into collection
collection.insert_many(NBA_College)
```

```python
# Define Schools Budget Collection
collection = db['Schools_Budget']

# Convert it to a dictionary
SchoolsBudget_dict = df3.to_dict('records')
# print(SchoolsBudget_dict)
# Inserting the dictionary into collection
collection.insert_many(SchoolsBudget_dict)
```

```python
# Define Tax Data Collection
collection = db['Tax_Data']

# Convert it to a dictionary
TaxData_dict = df4.to_dict('records')
# print(TaxData_dict)
# Inserting the dictionary into collection
collection.insert_many(TaxData_dict)
```
